```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Cargar datos desde un archivo CSV
filename = "clusters.csv"  # Cambia esto al nombre de tu archivo
data = pd.read_csv(filename, header=None)

# Separar características y etiquetas de clúster
features = data.iloc[:, :-1]  # Todas las columnas excepto la última
clusters = data.iloc[:, -1]  # Última columna: ClusterID

# --- Visualización 2D (usando las dos primeras dimensiones) ---
def plot_2d(features, clusters):
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.7)
    plt.colorbar(scatter, label="ClusterID")
    plt.xlabel("dim 1")
    plt.ylabel("dim 2")
    plt.title("Cluster Visualization (2D)")
    plt.show()

# --- Visualización 3D (usando las tres primeras dimensiones) ---
def plot_3d(features, clusters):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(features.iloc[:, 0], features.iloc[:, 1], features.iloc[:, 2], c=clusters, cmap='viridis', s=50, alpha=0.7)
    ax.set_xlabel("dim 1")
    ax.set_ylabel("dim 2")
    ax.set_zlabel("dim 3")
    plt.title("Cluster Visualization (3D)")
    plt.colorbar(scatter, label="ClusterID")
    plt.show()

# --- Visualización PCA (reducción a 2 dimensiones) ---
def plot_pca(features, clusters):
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(features)
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.7)
    plt.colorbar(scatter, label="ClusterID")
    plt.xlabel("PCA dim 1")
    plt.ylabel("PCA dim 2")
    plt.title("Cluster Visualization with PCA (2D)")
    plt.show()

# --- Visualización t-SNE (reducción a 2 dimensiones) ---
def plot_tsne(features, clusters):
    tsne = TSNE(n_components=2, random_state=42)
    reduced_data = tsne.fit_transform(features)
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.7)
    plt.colorbar(scatter, label="ClusterID")
    plt.xlabel("t-SNE dim 1")
    plt.ylabel("t-SNE dim 2")
    plt.title("Cluster Visualization with t-SNE (2D)")
    plt.show()

# Llamar a cada función para visualizar
print("Visualización 2D:")
plot_2d(features, clusters)

print("Visualización 3D:")
plot_3d(features, clusters)

print("Visualización con PCA:")
plot_pca(features, clusters)

print("Visualización con t-SNE:")
plot_tsne(features, clusters)
```
